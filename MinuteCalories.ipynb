```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load datasets
df=pd.read_csv("/content/minuteCaloriesWide_merged.csv")
df1=pd.read_csv("/content/minuteCaloriesNarrow_merged.csv")
```

```python
df.head()
```

```python
df1.head()
```

```python
#Splitting ActivityMinute and ActivityHour from df,df1 with Date and Time
df1[['Date','Time']]=df1['ActivityMinute'].str.split(' ', 1, expand=True)
df[['Date','Time']]=df['ActivityHour'].str.split(' ', 1 , expand=True)

# Summing up 'calories00' to 'calories59' and creating a new column 'total_steps'
df['total_calories'] = df.loc[:, 'Calories00':'Calories59'].sum(axis=1)

# Dropping columns 'calories00' to 'calories59'
df.drop(columns=df.loc[:, 'Calories00':'Calories59'], inplace=True)
```

```python
#Dropping unnecessary column from df1 and df
df1.drop('ActivityMinute',axis=1,inplace=True)
```

```python
df.drop('ActivityHour',axis=1,inplace=True)
```

```python
# Merge df and df1 based on 'Id'
merged_df = pd.merge(df, df1, on=('Id','Date'))
merged_df.sample()
```

```python
#Add two columns of calories
merged_df['MinuteCalories'] = merged_df['total_calories'] + merged_df['Calories']
#Drop the unnecessary columns
merged_df.drop(['total_calories','Calories'],axis=1,inplace=True)
merged_df.drop(['Time_x','Time_y'],axis=1,inplace=True)
```
