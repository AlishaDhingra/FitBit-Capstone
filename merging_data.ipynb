**merging weight data with merge dataset**

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/weightLogInfo_merged.csv")
```

```python
df.head()
```

```python
# Rename the 'Date' column to 'date'
df.rename(columns={'Date': 'date'}, inplace=True)

#splitting date in Date and Time
df[['Date','Time']]=df['date'].str.split(' ', 1, expand=True)

#Dropping unnecessary columns
df.drop(['date','WeightPounds'],axis=1,inplace=True)

df.sample()
```

```python
df1=pd.read_csv("/content/sleepDay_merged.csv")
```

```python
df1.head()
```

```python
df2=pd.read_csv("/content/minuteSleep_merged.csv")
```

```python
df2.head()
```

```python
#Splitting sleepday and date with Date and Time of df1,df2 dataset
df1[['Date','Time']]=df1['SleepDay'].str.split(' ',1,expand=True)
df2[['Date','Time']]=df2['date'].str.split(' ', 1, expand=True)

#Dropping the unnecessary columns
df1.drop('SleepDay',axis=1,inplace=True)
df2.drop('date',axis=1,inplace=True)

#merging data df1,df2
merged_df=pd.merge(df1,df2, on=('Date','Id'))
```

```python
#merge data df and merged_df
merged_df1=pd.merge(df,merged_df,on = ('Date','Id'))

#Dropping the unnecessary columns
merged_df1.drop(['Time_x','Time_y'],axis=1,inplace=True)
merged_df1.sample()
```

```python
# Save the DataFrame as a CSV file
merged_df1.to_csv('df1.csv', index=False)

from google.colab import files

# Download the CSV file
files.download('df1.csv')
```
