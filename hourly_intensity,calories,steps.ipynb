```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/hourlySteps_merged.csv")
```

```python
df.head()
```

```python
df1=pd.read_csv("/content/hourlyIntensities_merged.csv")
```

```python
df1.head()
```

```python
#Splitting ActivityHour in df,df1 with Date and Time
df[['Date','Time']]=df['ActivityHour'].str.split(' ',1,expand=True)
df1[['Date' , 'Time']]=df1['ActivityHour'].str.split(' ',1, expand=True)

#Drop unnecessary columns from df,df1
df.drop('ActivityHour',axis=1,inplace=True)
df1.drop('ActivityHour',axis=1,inplace=True)
```

```python
#merging data of df,df1
merged_df=pd.merge(df,df1,on=('Date','Id'))

#Dropp unnecessary columns
merged_df.drop(['Time_x','Time_y'],axis=1,inplace=True)
merged_df.sample()
```
