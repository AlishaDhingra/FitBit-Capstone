```python
import pandas as pd
```

```python
df=pd.read_csv("/content/dailyActivity_merged.csv")
df1=pd.read_csv("/content/heartrate_seconds_merged.csv")
```

```python
df1.head()
```

```python
# Rename the 'ActivityDate' column to 'Date'
df.rename(columns={'ActivityDate': 'Date'}, inplace=True)
df.sample()
```

```python
#Splitting Time in Date and time
df1[['Date','time']]=df1['Time'].str.split(' ', 1, expand=True)

#Dropping unnecessary columns
df1.drop('Time',axis=1,inplace=True)

# Calculate the average heart rate
df1['average_heart_rate'] = df1['Value'].mean()
df1.sample()
```

```python
#merging df1 with df
merged_df=pd.merge(df,df1,on=('Date','Id'))
merged_df.sample()
```

```python
# Save the DataFrame as a CSV file
merged_df.to_csv('merged.csv', index=False)
```
