```python
import matplotlib.pyplot as plt
# Calculate total weight for each Id
total_weight = df.groupby('Id')['WeightKg'].sum()

# Plot pie chart for weight by unique Id
plt.figure(figsize=(8, 6))
plt.pie(total_weight, labels=total_weight.index, autopct='%1.1f%%', startangle=140)
plt.title('Weight Distribution by Unique Id')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import seaborn as sns
# Plot bar graph for BMI and Fat for each unique Id
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Id', y='BMI', color='blue', label='BMI')
plt.xlabel('Id')
plt.ylabel('Value')
plt.title('BMI by Unique Id')
plt.legend()
plt.show()
```

```python
# Plot sleep records for each unique Id
plt.figure(figsize=(10, 6))
for id, group in df.groupby('Id'):
    plt.plot([id]*len(group), group['TotalMinutesAsleep'], 'o', label=f'Id {id}')

plt.xlabel('Id')
plt.ylabel('Sleep Duration (minutes)')
plt.title('Sleep Records by Unique Id')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Use Seaborn for a grouped bar chart
sns.barplot(x='Date', y='VeryActiveMinutes', data=df1, color='black', label='Very Active Minutes', alpha=0.7)
sns.barplot(x='Date', y='LightlyActiveMinutes', data=df1, color='green', label='Lightly Active Minutes', alpha=0.7)

# Adding labels and title
plt.title('Very Active Minutes vs Lightly Active Minutes per Day')
plt.xlabel('Activity Day')
plt.ylabel('Minutes')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(30, 8))
colors = ['pink' if active else 'lightgreen' for active in df1['VeryActiveDistance']]
plt.scatter(df1['VeryActiveDistance'], df1['LightActiveDistance'], color=colors, alpha=0.5)
plt.title('Very Active Distance vs Light Active Distance')
plt.xlabel('Very Active Distance')
plt.ylabel('Light Active Distance')
plt.grid(True)
plt.show()
```

```python
# Group by ActivityDay and calculate the total distance per day
distance_per_day = df1.groupby('Date')['TotalDistance'].sum()

# Plotting the line chart for distance per day
plt.figure(figsize=(30, 5))
plt.plot(distance_per_day.index, distance_per_day.values, marker='o', linestyle='-', color='green')
plt.title('Total Distance Per Day')
plt.xlabel('Activity Day')
plt.ylabel('Total Distance (in miles)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot bar graph for  heart rate by unique Id
plt.figure(figsize=(10, 6))
sns.barplot(data=df1, x='Id', y='Value', color='blue')
plt.xlabel('Id')
plt.ylabel(' Heart Rate')
plt.title(' Heart Rate by Unique Id')
plt.show()
```

```python
# Calories Per Day Intervals for each user .
# Group by 'TimeOfDay' and 'Id' and calculate total calories for each time interval and each user
calories_per_time_of_day_user = df1.groupby(['Id', 'time'])['Calories'].sum().unstack()

# Plotting the bar graph for each user
fig, axes = plt.subplots(nrows=len(calories_per_time_of_day_user), figsize=(30, 5 * len(calories_per_time_of_day_user)))

for i, (user_id, calories_per_user) in enumerate(calories_per_time_of_day_user.iterrows()):
    calories_per_user.plot(kind='bar', ax=axes[i], color=['pink', 'green', 'blue'])
    axes[i].set_xlabel('Time of Day')
    axes[i].set_ylabel('Total Calories')
    axes[i].set_title(f'Total Calories Per Time of Day - User {user_id}')
    axes[i].tick_params(axis='x', rotation=0)

plt.show()
```

```python
# Total Steps Per Day Intervals for each user
# Group by 'TimeOfDay' and 'Id' and calculate total steps for each time interval and each user
steps_per_time_of_day_user = df1.groupby(['Id', 'time'])['TotalSteps'].sum().unstack()

# Plotting the line chart for each user
fig, axes = plt.subplots(nrows=len(steps_per_time_of_day_user), figsize=(30, 5 * len(steps_per_time_of_day_user)))

for i, (user_id, steps_per_user) in enumerate(steps_per_time_of_day_user.iterrows()):
    steps_per_user.plot(kind='line', marker='o', ax=axes[i])
    axes[i].set_xlabel('Activity Hour')
    axes[i].set_ylabel('Total Steps')
    axes[i].set_title(f'Total Steps Per Hour - User {user_id}')
    axes[i].legend(title='Time of Day')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# Plotting TotalMinutesAsleep and TotalTimeInBed as a bar graph
plt.figure(figsize=(30,10))

# Bar width can be adjusted based on your preference
bar_width = 0.35

# Bar positions
bar_positions1 = df['Date']
bar_positions2 = bar_positions1 + pd.Timedelta(days=bar_width)

# Bar heights
plt.bar(bar_positions1, df['TotalMinutesAsleep'], width=bar_width, label='Total Minutes Asleep')
plt.bar(bar_positions2, df['TotalTimeInBed'], width=bar_width, label='Total Time In Bed')

plt.title('Sleep Patterns Over Time (Bar Graph)')
plt.xlabel('Sleep Day')
plt.ylabel('Minutes')
plt.legend()
plt.show()
```

```python
# Assuming 'Date' column is stored as string or object data type
# Convert 'Date' column to datetime data type
df1['Date'] = pd.to_datetime(df1['Date'])

# Now you can use .dt accessor to extract day of the week
df1["DayOfTheWeek"] = df1["Date"].dt.day_name()
df1
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6)) # sepcify size of the chart
plt.hist(df1.DayOfTheWeek, bins = 7, width = 0.6, edgecolor = "black")

# adding annotations and visuals

plt.xlabel("Day of the week")
plt.ylabel("Frequency")
plt.title("No. of times users logged in app across the week")
plt.grid(True)
plt.show()
```

```python
very_active_mins = df1["VeryActiveMinutes"].sum()
fairly_active_mins = df1["FairlyActiveMinutes"].sum()
lightly_active_mins = df1["LightlyActiveMinutes"].sum()
sedentary_mins = df1["SedentaryMinutes"].sum()


slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
colours = ["coral", "green", "skyblue", "orange"]
explode = [0, 0, 0, 0.1]
plt.style.use("default")
plt.pie(slices, labels = labels,
        colors = colours, wedgeprops = {"edgecolor": "black"},
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```
