```python
#importing libraraies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df=pd.read_csv("/content/minuteIntensitiesWide_merged.csv")
```

```python
df.head()
```

```python
# Summing up 'intensitie00' to 'intensity59' and creating a new column 'total_steps'
df['total_intensities'] = df.loc[:, 'Intensity00':'Intensity59'].sum(axis=1)

# Dropping columns 'intensity00' to 'intensity59'
df.drop(columns=df.iloc[:, 2:62], inplace=True)
df.sample()
```

```python
#Splitting 'ActivityHour' in 'Date' and 'Time' columns.
df[['Date','Time']]=df['ActivityHour'].str.split(' ',1,expand=True)

#Drop unnecessary column
df.drop('ActivityHour',axis=1,inplace=True)
df.sample()
```

```python
#load dataset
df1=pd.read_csv("/content/minuteIntensitiesNarrow_merged.csv")
```

```python
df1.head()
```

```python
#Splitting ActivityMinute in Date and Time columns
df1[['Date','Time']]=df1['ActivityMinute'].str.split(' ',1,expand=True)
#Dropping unnecessary column
df1.drop('ActivityMinute',axis=1,inplace=True)
df1.sample()
```

```python
# Merge DataFrames based on 'Date' and 'Id'
merged_df = pd.merge(df, df1, on=['Date', 'Id'])


# Add two columns together and store the result in a new column
merged_df['Minuteintensities'] = merged_df['total_intensities'] + merged_df['Intensity']

#Drop unnecessary columns
merged_df.drop('total_intensities',axis=1,inplace=True)
merged_df.drop('Intensity',axis=1,inplace=True)
#Dropping Time columns
merged_df.drop('Time_x',axis=1,inplace=True)
merged_df.drop('Time_y',axis=1,inplace=True)
merged_df.sample()
```
